```python
import pandas as pd

# 1. clean datasheet

df = pd.read_excel("HLA_freq.xlsx", ["A", "B", "C", "DPA1", "DPB1", "DQA1", "DQB1", "DRB1"])
```

```python
# cleaned_data = {}
with pd.ExcelWriter("max_freq.xlsx") as fh:

    for key, sheet in df.items():

        a = sheet.groupby(["Allele", "Population"])["Allele Frequency"].sum() # add up frequency column
        b = sheet.groupby(["Allele", "Population"])["Sample Size"].mean() # keep population column
        df_clean = pd.DataFrame([a,b]).T
        max_freq = df_clean.groupby("Allele")["Allele Frequency"].max()

        max_freq.to_excel(fh, sheet_name=key)
        # cleaned_data[key] = max_freq
```
