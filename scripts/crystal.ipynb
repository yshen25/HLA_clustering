```python
def validate(InDir):
    # blast against all allele sequences, and find best match
    result = []
    for InFas in os.listdir(InDir):
        if InFas.endswith('.fasta'):
            # print(InFas)
            blast_result = blastp(f"{InDir}/{InFas}")
            result.extend(blast_result)
            top_pident = float(blast_result[0][2])/float(blast_result[0][3]) + float(blast_result[5][2])/float(blast_result[5][3])
            if top_pident < 2:
                print(InFas)
            # blast_result = pd.DataFrame(blast_result, columns=["Allele", "nident", "qlen"])
            # blast_result = blast_result.astype({"Allele":"object", "nident":"float64", "qlen":"float64"})
            # blast_result['pident'] = blast_result['nident']/blast_result['qlen']
            # top_pident = blast_result["pident"].iloc[0] + blast_result["pident"].iloc[5]
            # blast_result = np.array(blast_result)
            # A_result = pd.DataFrame(blast_result[0:5], dtype=(str, float, float))
            # A_result = blast_result.iloc[0:5]
            
            # print(A_result['nident']/A_result['qlen'])
            # print(A_result.iloc[:,1]/A_result.iloc[:,2])
            # A_result = np.hstack((A_result, A_result[:,1]/A_result[:,2]))
            # B_result = blast_result.iloc[5:10]
            # print(top_pident, A_result, B_result)
            # B_result = np.hstack((B_result, B_result[:,1]/B_result[:,2]))

    result = pd.DataFrame(result, columns=["file", "Allele", "nident", "qlen", "gaps", "sstart", "send"])
    result = result.astype({"file": "object","Allele":"object", "nident":"float64", "qlen":"float64"})
    result['pident'] = result['nident']/result['qlen']

    return result
    

result = validate("../crystal/Class2")
result.to_csv("HLA2_crystal.csv")
```
